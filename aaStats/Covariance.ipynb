```python
import ceo
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
NL = 48
n = NL
NA = NL+1
D = 25.5
d = D/NL
```

```python
atm = ceo.GmtAtmosphere(0.15,25)
ngs = ceo.Source("V",resolution=(NA,NA))
```

# Angle of arrival covariance

```python
aa = ceo.AaStatsMatrix(NL,atm,d,ngs)

fig,ax = plt.subplots(figsize=(10,4))
h = ax.matshow(np.hstack(np.vsplit(aa.cov.host(),4)))
fig.colorbar(h,ax=ax,orientation='horizontal')
```

# Phase/Angle of arrival covariance

```python
d = D/(NA-1)
pa = ceo.PaStats(NA,NL,1,atm,d,ngs,ngs)

fig,ax = plt.subplots(figsize=(6,4))
h = ax.matshow(np.hstack(np.vsplit(pa.cov.host().reshape(-1,NA + NL -1),2)))
fig.colorbar(h,ax=ax,orientation='horizontal')
```

```python
osf = 4
NP = osf*NL+1
pa = ceo.PaStats(NP,NL,osf,atm,d,ngs,ngs)

fig,ax = plt.subplots(figsize=(6,4))
h = ax.matshow(np.hstack(np.vsplit(pa.cov.host().reshape(-1,NP + NL -1),2)))
fig.colorbar(h,ax=ax,orientation='horizontal')
```

```python
pa = ceo.APaStats(NA,NL,1,atm,d,ngs,0)

fig,ax = plt.subplots(figsize=(6,4))
h = ax.matshow(np.hstack(np.vsplit(pa.cov.host().reshape(-1,NA + NL -1),2)))
fig.colorbar(h,ax=ax,orientation='horizontal')
```

```python
pa = ceo.APaStats(NA,NL,1,atm,d,ngs,ceo.constants.ARCMIN2RAD*5)

fig,ax = plt.subplots(figsize=(6,4))
h = ax.matshow(np.hstack(np.vsplit(pa.cov.host().reshape(-1,NA + NL -1),2)))
fig.colorbar(h,ax=ax,orientation='horizontal')
```
